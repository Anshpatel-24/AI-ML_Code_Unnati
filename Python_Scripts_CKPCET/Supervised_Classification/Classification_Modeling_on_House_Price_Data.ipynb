## Prepare Data for Machine Learning.

```python
# Seperate the target variable and independent variable.
target = df_house_price['Sold']
```

```python
# Independent variable
df_independant = df_house_price.drop('Sold', axis = 1)
```

## Seperate the Categorical Columns from Independent Variables

```python
df_cat = df_independant.select_dtypes(include = 'object')
```

```python
df_cat.head()
```

```python
# Encode the categorical variables
encoded = pd.get_dummies(df_cat)
encoded = encoded.reset_index()
```

```python
encoded.head()
```

## Scale the Numerical Data

```python
df_num = df_independant.select_dtypes(include = 'number')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initiate the StandardScalar

ss = StandardScaler()
```

```python
# Transform the data

scale = ss.fit_transform(df_num)
```

```python
print(scale)
```

```python
df_scale = pd.DataFrame(scale, columns = df_num.columns)
```

```python
df_scale = df_scale.reset_index()
df_scale
```

```python
# Concate the dummy variables with numeric features to create a dataframe of
# 'axis = 1'concate the dataframes along columns

X = pd.concat([df_scale, encoded], axis = 1)

# Display First Five Observation
X.head()
```

```python
# import train_test_split
# Split the data for Training and Testing
# Import various function from statsmodels

from sklearn.model_selection import train_test_split
import statsmodels
import statsmodels.api as sm

X = sm.add_constant(X)

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   target,
                                                   random_state = 10,
                                                   test_size = 0.2)
print('X_train', X_train.shape)
print('y_train', y_train.shape)
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Decision Tree Classifier
tree_model = DecisionTreeClassifier()

# Fit the model to the training data
tree_model.fit(X_train, y_train)

# Predict on the Testing data
predictions = tree_model.predict(X_test)
```

```python
predictions
```

```python
# Calculating the accuracy

accuracy = accuracy_score(y_test, predictions)
print("Accuracy :", accuracy)
```
